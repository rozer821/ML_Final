```python
import math
import pandas as pd
import numpy as np
df = pd.read_csv("qudditch_training.csv", header=0)
```

```python
df.columns
```

```python
df['house'].replace({'?':0,'Gryffindor':1,'Slytherin':2,'Ravenclaw':3,'Hufflepuff':4,'Other':5},inplace = True)

df['gender'].replace({'Unknown/Invalid':0,'Female':1,'Male':2},inplace = True)

df['weight'].replace({'?':0,'[0-25)':1,'[25-50)':2,'[50-75)':3,'[75-100)':4,
                      '[100-125)':5,'[125-150)':6, '[150-175)':7,'[175-200)':8,'>200':9},inplace = True)

df['player_code'].replace({'?':0, 'MC':1, 'MD':2, 'HM':3, 'UN':4, 'BC':5, 'SP':6, 'CP':7, 'SI':8, 'DM':9, 'CM':10,
       'CH':11, 'PO':12, 'WC':13, 'OT':14, 'OG':15, 'MP':16, 'FR':17},inplace = True)

df['move_specialty'] = df['move_specialty'].str.replace('specialty','')
df['move_specialty'].replace({'?':0},inplace = True)

df['player_type'].replace({'Beater1':0, 'Beater2':1, 'Multiple':2, 'Captain':3, 'Chaser1':4, 'Chaser2':5, 'Chaser3':6,
       'Keeper':7, 'Seeker':8},inplace = True)

df['snitchnip'].replace({'None':0, 'Norm':1,'>200':2, '>300':3},inplace = True)

df['stooging'].replace({'None':0, 'Norm':1, '>7':2, '>8':3},inplace = True)

df['body_blow'].replace({'No':0, 'Steady':1, 'Up':3, 'Down':4},inplace = True)

df['checking'].replace({'No':0, 'Steady':1, 'Up':3, 'Down':4},inplace = True)

df['dopplebeater_defence'].replace({'No':0, 'Steady':1, 'Up':3, 'Down':4},inplace = True)

df['hawkshead_attacking_formation'].replace({'No':0, 'Steady':1, 'Up':3, 'Down':4},inplace = True)

df['no_hands_tackle'].replace({'No':0, 'Steady':1, 'Up':3, 'Down':4},inplace = True)

df['power_play'].replace({'No':0, 'Steady':1},inplace = True)

df['sloth_grip_roll'].replace({'No':0, 'Steady':1, 'Up':3, 'Down':4},inplace = True)

df['spiral_dive'].replace({'No':0, 'Steady':1, 'Up':3, 'Down':4},inplace = True)

df['starfish_and_stick'].replace({'No':0, 'Steady':1},inplace = True)

df['twirl'].replace({'No':0, 'Steady':1, 'Up':3, 'Down':4},inplace = True)

df['wronski_feint'].replace({'No':0, 'Steady':1, 'Up':3, 'Down':4},inplace = True)

df['zig-zag'].replace({'No':0, 'Steady':1, 'Up':3, 'Down':4},inplace = True)

df['bludger_backbeat'].replace({'No':0, 'Steady':1, 'Up':3, 'Down':4},inplace = True)

df['chelmondiston_charge'].replace({'No':0, 'Steady':1},inplace = True)

df['dionysus_dive'].replace({'No':0, 'Steady':1,'Up':3},inplace = True)

df = df.drop(columns=['double_eight_loop', 'finbourgh_flick'])

df['reverse_pass'].replace({'No':0, 'Steady':1, 'Up':3, 'Down':4},inplace = True)

df['parkins_pincer'].replace({'No':0, 'Steady':1, 'Up':3, 'Down':4},inplace = True)

df['plumpton_pass'].replace({'No':0, 'Steady':1},inplace = True)

df['porskoff_ploy'].replace({'No':0, 'Steady':1},inplace = True)

df['transylvanian_tackle'].replace({'No':0, 'Steady':1},inplace = True)

df['woollongong_shimmy'].replace({'No':0, 'Steady':1},inplace = True)

df['change'].replace({'No':0, 'Ch':1},inplace = True)

df['snitch_caught'].replace({'No':0, 'Yes':1},inplace = True)

df['quidditch_league_player'].replace({'NO':0, 'YES':1},inplace = True)
```
